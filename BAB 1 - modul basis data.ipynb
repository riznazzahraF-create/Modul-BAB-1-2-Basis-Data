## **Entitas**
Entitas adalah suatu objek atau konsep yang memiliki keberadaan sendiri dan dapat dibedakan dari objek atau konsep lainnya. Entitas dapat berupa benda, orang, tempat, atau konsep abstrak.


## **Atribut**
Atribut adalah sifat atau karakteristik yang dimiliki oleh suatu entitas atau objek. Atribut digunakan untuk menggambarkan atau mendefinisikan entitas tersebut.

## **Primary Key (PK)**
Primary key adalah suatu atribut atau kombinasi atribut yang digunakan untuk mengidentifikasi secara unik setiap record atau baris data dalam sebuah tabel database.

## **Foreign Key (FK)**
Foreign key adalah suatu atribut atau kombinasi atribut dalam sebuah tabel database yang merujuk ke primary key di tabel lain.

---



### **1. Entitas Kuat** *(Strong Entity)*

* Setiap entitas kuat â†’ satu tabel.
* Setiap atribut sederhana â†’ kolom.
* Atribut kunci â†’ Primary Key.

---

### **2. Composite Attribute**

* Entitas tetap menjadi satu tabel.
* Atribut sederhana â†’ kolom.
* Atribut yang merupakan komposit â†’ tidak dijadikan kolom langsung (dipecah atau tidak dipakai).

---

### **3. Multivalue Attribute**

* Atribut multivalue â†’ tabel baru.
* Maka entitas kuat + multivalue â†’ menjadi dua tabel.

---

### **4. Derived Attribute**

* Atribut turunan â†’ jadi kolom (opsional sesuai kebutuhan sistem).

---

### **5. Entitas Lemah** *(Weak Entity)*

* Menjadi tabel terpisah.
* Semua atribut sederhana â†’ kolom.
* PK biasanya gabungan partial key + FK dari entitas kuat.

---

### **6. Relasi One-to-One** (1:1)

* Primary key salah satu entitas â†’ foreign key di entitas lain.

---

### **7. Relasi One-to-Many** (1:M)

* Primary key entitas â€œoneâ€ â†’ FK pada entitas â€œmanyâ€.

---

### **8. One-to-Many** (1:M) **dengan Atribut Relasi**

* Dibuat 3 tabel: entitas 1, entitas 2, dan tabel relasi.
* Tabel relasi berisi: PK/FK dari kedua entitas + atribut relasi.

---

### **9. Relasi Many-to-Many** (M:N)

* Relasi â†’ tabel baru.
* Tabel relasi berisi FK dari kedua entitas + atribut sederhana relasinya.

---

### **10. Relasi Unary**  *(Recursive Relationship)*

* Satu tabel saja.
* Tambahkan FK dalam tabel yang sama untuk menunjukkan relasi diri sendiri.

---

### **11. Relasi Ternary**

* Relasi â†’ tabel baru.
* Atribut sederhana â†’ kolom.
* FK dari ketiga entitas masuk ke tabel relasi.

---

### **12. GENSPEC** *(Generalization & Specialization)*

*Metode 1* (Superclass & Subclass terpisah):

* Superclass â†’ tabel sendiri (PK + atribut sederhana).
* Subclass â†’ tabel sendiri yang mewarisi PK superclass.

*Metode 2* (Hanya Subclass):

* Setiap subclass â†’ tabel sendiri.
* Semua atribut superclass + subclass â†’ ada di tabel subclass.
* PK diambil dari superclass.

---

### **13. Agregasi**

* Tabel agregasi menggunakan PK dari entitas agregat.
* Relasi ke entitas lain menggunakan FK dari setiap entitas yang terlibat.
