###**3.3 SQL**
3 bagian perintah dalam SQL:

* **DDL** *(Data Definition Language)*:

Perintah SQL yang berkaitan dengan pendefinisian suatu struktur database (meliputi database dan tabel). Perintahnya: CREATE, ALTER, RENAME, dan DROP.

* **DML** *(Data Manipulation Language)*:

Perintah SQL yang berkaitan dengan manipulasi atau pengolahan data (record) di dalam tabel. Perintahnya: SELECT, INSERT, UPDATE, dan DELETE.

* **DCL** *(Data Control Language)*:

Perintah SQL yang berkaitan dengan manipulasi user dan hak akses (privileges). Perintahnya: GRANT dan REVOKE.

### **3.5 CONSTRAINT TABLE**
Constraint adalah aturan atau batasan yang diterapkan pada tabel untuk menjaga integritas dan konsistensi data.

Ada 5 aturan constraint yang umum diterapkan, baik saat membuat tabel (CREATE TABLE) maupun memodifikasi tabel (ALTER TABLE):

* PRIMARY KEY
* FOREIGN KEY
* UNIQUE
* NOT NULL
* CHECK

Dalam penerapan skema relasi, tabel harus dibuat secara berurutan, dimulai dari tabel kuat (hasil entitas tunggal yang tidak memiliki Foreign Key), baru kemudian dilanjutkan ke tabel hasil relasi.

### **3.6 NILAI OTOMATIS DAN NILAI DEFAULT**

*3.6.1 Nilai otomatis / Auto Increment*

Auto Increment adalah mekanisme yang secara otomatis memberikan nilai pada suatu field (biasanya Primary Key), di mana nilai tersebut akan bertambah 1 setiap kali ada penambahan data baru.

*3.6.2 Nilai default*

Nilai default adalah pemberian nilai secara otomatis oleh sistem terhadap field tertentu (umumnya bernilai NULL) jika data tidak diisi. Perintahnya adalah Default Nilai_Default.

*3.7 TYPE : INNODB DAN XTRADB*

InnoDB: Merupakan storage engine yang sering digunakan di MySQL, terutama di lingkungan web. Fitur utamanya adalah dukungan transaksi (commit, rollback, dan crash recovery) serta fitur relasi tabel dan integritas melalui Foreign Key. InnoDB menjadi default table engine sejak MySQL versi 5.5.

XtraDB: Merupakan pengganti InnoDB di MariaDB dan dianggap lebih stabil.

Untuk menggunakan tipe mesin tertentu, sintaks Engine = InnoDB; dapat ditambahkan di akhir perintah CREATE TABLE.
